```python
from tkinter import *
```

```python
from tkinter.ttk import Frame, Notebook
```

```python
window = Tk()

window.title('Note Pad')

window.geometry ('500x500')



menu = Menu(window)



newitem1 = Menu(menu, tearoff =0)

newitem1.add_command (label = 'New')

newitem1.add_command (label = 'New window')

newitem1.add_command (label = 'Open')

newitem1.add_command (label = 'Save')

newitem1.add_command (label = 'Save as')

newitem1.add_separator()

newitem1.add_command (label = 'Page setup')

newitem1.add_command (label = 'Print')

newitem1.add_separator()

newitem1.add_command (label = 'Exit')


newitem2 = Menu(menu, tearoff =0)

newitem2.add_command (label = 'Undo')

newitem2.add_command (label = 'Cut')

newitem2.add_command (label = 'Copy')

newitem2.add_command (label = 'Paste')

newitem2.add_command (label = 'Delete')

newitem2.add_separator()

newitem2.add_command (label = 'Search with Bing...')

newitem2.add_command (label = 'Find')

newitem2.add_command (label = 'Find Next')

newitem2.add_command (label = 'Find Previous')

newitem2.add_command (label = 'Replace')
                      
newitem2.add_command (label = 'Go To..')

newitem2.add_separator()

newitem2.add_command (label = 'Select All')

newitem2.add_command (label = 'Time/ Date')


newitem3 = Menu(menu, tearoff =0)

newitem3.add_command (label = 'Word Wrap')

newitem3.add_command (label = 'Font')







view_menu = Menu(menu, tearoff =0)

zoom_menu =Menu(view_menu, tearoff =0)

def zoon_in():
    print('zoom in')

def zoom_out():
    print('zoom out')
    

zoom_menu.add_command (label = 'Zoom in', command = zoon_in)

zoom_menu.add_command (label = 'zoom out', command = zoom_out)

view_menu.add_cascade (label ='zoom', menu= zoom_menu)

view_menu.add_command (label ='Status Bar', command = lambda : print('Status Bar'))




newitem5 = Menu(menu, tearoff =0)

newitem5.add_command (label = 'View help')

newitem5.add_command (label = 'Send Feedback')

newitem5.add_separator()

newitem5.add_command (label = 'About Notepad')




                      







menu.add_cascade (label ='File', menu = newitem1)
menu.add_cascade (label ='Edit', menu =newitem2)
menu.add_cascade (label ='Format', menu =newitem3)
menu.add_cascade (label ='View', menu = view_menu)
menu.add_cascade (label ='Help', menu =newitem5)


text_area =Text(window, wrap='word', undo=True)
text_area.pack(expand=True, fill='both')






window.configure(menu = menu)
























window.mainloop()
```
