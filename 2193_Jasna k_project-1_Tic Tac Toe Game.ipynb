```python
from IPython.display import clear_output


def print_board(board):# printing game board
    clear_output()
    for raw in board:
        print(" | ".join(raw))
        print("-" *9)

def check_winner(board,player):
    for raw in board:
        if all(cell==player for cell in raw):#checking the winner
            return True

    for col in range(3):
         if all(board[raw][col]==player for raw in range(3)):
              return True

    if all(board[i][i]==player for i in range(3)) or all(board[i][2 - i]==player for i in range(3)):
        return True

    return False


def is_full(board):
    return all(cell in (' X ',' O ')for raw in board for cell in raw)

def full_board_check(board):
    return is_full(board)


def player_choice(board):
    while True:
        try:
            position = int(input("Enter your move 1-9: "))-1
            raw, col = divmod(position ,3)
            if board[raw][col]== ' ':
                return raw, col
            else:
                print("that spot is already taken. try again")
        except (valueError , IndexError):
            print("Invalid input. Enter a number betwean 1-9.")

def replay():#move to replay the game
    return input("Do you want to play again? (Y/N): ").lower() == 'y'

def tic_tac_toe():
    while True:
        board = [[' ' for _ in range(3)] for _ in range(3)]
        players = ['X', 'O']


        print("Welcome to tic tac toe")
        print_board(board)

        while True:
            for player in players:
                print(f"player{player}'s turn.")
                raw,col = player_choice(board)
                board[raw][col]= player
                print_board(board)
                
                if check_winner(board, player):
                    print(f"player{player} wins!")
                    break
                    
                if full_board_check(board):
                    print("it is a draw")
                    break
                else:
                    continue
                break

        if not replay():
            break

tic_tac_toe()
            
```
